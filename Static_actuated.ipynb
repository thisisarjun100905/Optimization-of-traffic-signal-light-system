```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_actuated = pd.read_excel("Result.xlsx")
df_Static = pd.read_excel('Static.xlsx')
```

```python
df_actuated.sample(10)
```

```python
df_Static.sample(10)
```

```python
df_actuated.describe()
```

```python
df_Static.describe()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df_Static['Time'] , df_Static['Delay'] , label = 'Static_case')
plt.plot(df_actuated['Time'] , df_actuated['Delay'] , label = 'Actuated_case')

# Adding labels and title
plt.xlabel('Time')
plt.ylabel('Delay Time')
plt.title('Delay_time v/s Time')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df_Static['Time'] , df_Static['Total Queue Length (PCU)'] , label = 'Static_case')
plt.plot(df_actuated['Time'] , df_actuated['Total Queue Length (PCU)'] , label = 'Actuated_case')

# Adding labels and title
plt.xlabel('Time')
plt.ylabel('Total Queue Length (PCU)')
plt.title('Total Queue Length (PCU) v/s Time')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
